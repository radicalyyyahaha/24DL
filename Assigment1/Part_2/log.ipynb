```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_training_curves(log_file):
    df = pd.read_csv(log_file)

    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(df['Step'], df['Train Loss'], label='Train Loss')
    plt.plot(df['Step'], df['Test Loss'], label='Test Loss')
    plt.xlabel('Steps')
    plt.ylabel('Loss')
    plt.title('Training and Testing Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(df['Step'], df['Train Accuracy'], label='Train Accuracy')
    plt.plot(df['Step'], df['Test Accuracy'], label='Test Accuracy')
    plt.xlabel('Steps')
    plt.ylabel('Accuracy')
    plt.title('Training and Testing Accuracy')
    plt.legend()

    # Show the plots
    plt.tight_layout()
    plt.show()


plot_training_curves('training_log.csv')
```
